```python
# 1. 요청
res = requests.get(url)
print(res.status_code) # 항상 확인을 해야한다.

# 2. 뷰티풀슾에 넣기

# 제목 : h3.tit_view
# 기자이름 : span.info_view span.txt_info : nth-child(1)
# 입력일시 : span.info_view span.txt_info : nth-child(2)
# 기사내용 : div#harmonyContainer


if res.status_code == 200:
    soup = BeautifulSoup(res.text)
    soup.select('h3.tit_view') # 여러개라면 부모로 제약을 걸면된다.
    # 아이디는 상관없지만 클래스라면 같은 것이 있는지 확인해야한다.
    
    title_tag = soup.select_one('h3.tit_view')
    reporter_tag = soup.select_one('span.info_view span.txt_info:nth-child(1)')
    input_date_tag = soup.select_one('span.info_view span.txt_info:nth-child(2)')
    news_tag = soup.select_one('div#harmonyContainer')
    
    print(title_tag.text.strip()) # 있든 없든 strip()은 기본 옵션처럼 넣어주자.
    print(reporter_tag.text.strip())
    print(input_date_tag) # 입력 안나오게 하고싶어. 날짜만 나오게 하고싶다. 그럼 이 밑에처럼 하면됨.
    print(input_date_tag.select_one('span').text.strip()) # input_date_tag.span 요것도 가능하다!
    print(news_tag.text.strip())
    
    # find 나 select 는 꼭 soup이 아니더라도 태그에도 사용할 수 있다.
    
else:
    print(url, '-요청 실패', res.status_code)
```

## 다음 랭킹 뉴스 목록
- URL
- 조회할 항목들의 selector
    - 제목
    - 기사 링크
    - 신문사
    - 기사 요약

```python
# 공통 : ul.list_news2  div.cont_thumb
# 제목 : strong > a : text
# 기사 링크 : strong > a : href 속성
# 신문사 : strong > span : text
# 기사 요약 : div.desc_thumb > span.link_txt : text


url = 'https://news.daum.net/ranking/popular'

res = requests.get(url)
if res.status_code == 200:    
    soup = BeautifulSoup(res.text)
    news_list = soup.select('ul.list_news2 div.cont_thumb')
    
    content_list = [] # 조회한 내용(50개 뉴스)을 저장할 리스트 (링크, 제목, 신문사, 요약)
    
    for news in news_list:
        link = news.select_one('strong > a').get('href')
        title = news.select_one('strong > a').text.strip()
        newspaper = news.select_one('strong > span').text.strip()
        summary = news.select_one('div.desc_thumb > span.link_txt').text.strip()
        
        content_list.append([link, title, newspaper, summary])
```

```python
# 랭킹 뉴스 목록 조회 함수

def get_daum_ranking_news(type='popular'):
    '''
    랭킹뉴스목록 조회 결과를 파일로 저장.
    [매개변수]
        type : popular-많이 본 뉴스, kkomkkom-열독률, bestreply-댓글
    '''
    
    base_url = 'https://news.daum.net/ranking/'
    
    url = parse.urljoin(base_url, type)
    
    # 요청
    res = requests.get(url)
    
    # 처리
    if res.status_code == 200:    
        soup = BeautifulSoup(res.text)
        news_list = soup.select('ul.list_news2 div.cont_thumb')
        print(len(news_list))

        content_list = [] # 조회한 내용(50개 뉴스)을 저장할 리스트 (링크, 제목, 신문사, 요약)

        for news in news_list:
            link = news.select_one('strong > a').get('href')
            title = news.select_one('strong > a').text.strip()
            newspaper = news.select_one('strong > span').text.strip()
            summary = news.select_one('div.desc_thumb > span.link_txt').text.strip()

            content_list.append([link, title, newspaper, summary])
            
        # csv 파일로 저장 => pandas를 이용해서 표(DataFrame)를 생성 후 파일로 저장.
        filename = '{}_{}_news_list.csv'.format(type, datetime.now().strftime('%Y-%m-%d'))
        
        df = pd.DataFrame(content_list, columns=['링크', '제목', '신문사', '요약'])
        df.to_csv(filename, index=False, encoding='UTF-8')
            
    else:
        raise Exception('요청 실패 - {}'.format(res.status_code))
```
